```python
import orjson as json
import pandas as pd
```

```python
from features import chat_features
```

```python
import tqdm
```

```python
train_features = []
for line in tqdm.tqdm(open('../data/train_matches.jsonl')):
    game = json.loads(line)
    train_features.append(chat_features(game))
```

```python
test_features = []
for line in tqdm.tqdm(open('../data/test_matches.jsonl')):
    game = json.loads(line)
    test_features.append(chat_features(game))
```

```python
train_features = pd.DataFrame(train_features)
test_features = pd.DataFrame(test_features)
```

```python
train_features.to_csv('../features/chat_train.csv', index=False)
test_features.to_csv('../features/chat_test.csv', index=False)
```
