```python
from features import game_features
```

```python
import tqdm
```

```python
test_features = []
for line in tqdm.tqdm(open('../data/test_matches.jsonl')):
    game = json.loads(line)
    test_features.append(game_features(game))
```

```python
test_df = pd.DataFrame(test_features)
```
