```python
import orjson as json
```

```python
unique_items = dict()
for file in ['../data/train_matches.jsonl', '../data/test_matches.jsonl']:
    for line in open(file):
        game = json.loads(line)
        for p in game['players']:
            for source in ['hero_inventory', 'hero_stash']:
                for thing in p[source]:
                    if thing['id'] not in unique_items:
                        unique_items[thing['id']] = 1
                    else:
                        unique_items[thing['id']] += 1
```

```python
def strange_items_mapper(item):
    m = {'ward_dispenser': 'ward_observer'}
    if item in m:
        return m[item]
    elif 'tango' in item:
        return 'tango'
    elif 'dagon' in item:
        return 'dagon'
    elif 'necronomicon' in item:
        return 'necronomicon'
    elif 'travel_boots' in item:
        return 'travel_boots'
    else:
        return item
```

```python
items_index = []
for item, v in unique_items.items():
    item_name = item[5:]
    if item_name not in items_df.index:
        item_2 = strange_items_mapper(item_name)
        if item_2 == item_name:
            print("Strange item:", item, v)
    else:
        items_index.append(item_name)
```
